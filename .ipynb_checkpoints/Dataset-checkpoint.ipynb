# Depth Histogram 

```python
# Import packeges
%matplotlib inline
import os
import cv2
import os.path as osp
import numpy as np
from matplotlib import pyplot as plt
import pyzed.sl as sl
from IPython.display import display, Image
```

```python
def draw_image_histogram(images, channels, color='k'):
    hist = cv2.calcHist(images, [channels], None, [256], [0, 256])
    plt.plot(hist, color=color)
    plt.xlim([0, 256])


def show_color_histogram(image):
    for i, col in enumerate(['b', 'g', 'r']):
        draw_image_histogram(image, [i], color=col)
```

```python
main_dir = ".\\dataset"
depth_dir = osp.join(main_dir, "depth_frame")


images = []
for img in os.listdir(depth_dir):
    if "png" in img:
        images.append(cv2.imread(osp.join(depth_dir, img), cv2.IMREAD_UNCHANGED))

draw_image_histogram(images, 0)
plt.show()
```

```python
depth_dir = osp.join(main_dir, "depth_frame")
left_dir = osp.join(main_dir, "left_image")
right_dir = osp.join(main_dir, "right_image")
display(Image(filename=osp.join(left_dir, "163.png"), retina=True))
display(Image(filename=osp.join(right_dir, "163.png"), retina=True))
display(Image(filename=osp.join(depth_dir, "163.png"), retina=True))
```
